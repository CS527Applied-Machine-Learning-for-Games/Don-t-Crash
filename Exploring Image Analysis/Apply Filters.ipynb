```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from skimage import io
from skimage.feature import hog
from skimage import data, color, exposure
```

```python
#%matplotlib inline

# Read in the image
image = cv2.imread('right_2020_09_12_12_57_10_043.jpg')

# Change color to RGB (from BGR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

### Canny Filter 

```python
canny_filter = cv2.Canny(gray, 250, 300)
f, ax1 = plt.subplots(1, 1, figsize=(10,10))

ax1.set_title('canny')
ax1.imshow(canny_filter, cmap='gray')
```

```python
features, hog_image = hog(gray, orientations=8, pixels_per_cell=(2,2),
                    cells_per_block=(50, 50), visualize=True)
f2, ax3 = plt.subplots(1, 1, figsize=(10,10))

ax3.set_title('HOG image')
ax3.imshow(hog_image, cmap='gray')
```

### Gaussian Blur Filter 

```python
gray_blur = cv2.GaussianBlur(gray, (9, 9), 0)
f2, ax2 = plt.subplots(1, 1, figsize=(10,10))

ax2.set_title('gaussian blur')
ax2.imshow(gray_blur, cmap='gray')
```
